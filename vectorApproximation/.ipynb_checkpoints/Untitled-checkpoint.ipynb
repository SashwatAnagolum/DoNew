### Introduction

This notebook presents a solution to task 1 of the QOSF mentorship screen test - given a quantum state $| \psi \rangle$ over 4 qubits, minimize the distance between the output state of a variational quantum circuit $|U(\theta)\rangle$ and $|\psi\rangle$, and plot a graph showing how the minimum distance changes as a function of the number of layers used in the variational circuit.

**Modifications**

The task asks us to minimize

$$\sum_i \left ( |U(\theta)\rangle_i - |\psi\rangle_i \right )^2$$

However, due to the complex nature of the entries of the two vectors, we could end up minimizing the sum of squared differences, while still generating a $|U(\theta)\rangle$ that is very different from $|\psi\rangle$. For example, we might have to approximate the 1 qubit state given below:

$$ \begin{bmatrix} -\frac{1}{2} + \frac{1}{2}i \\\frac{1}{2} - \frac{1}{2}i \end{bmatrix} $$

At some point during the optimization process, $|U(\theta)\rangle$ might look like this:

$$ \begin{bmatrix} \frac{1}{2} + \frac{1}{2}i \\\frac{1}{2} + \frac{1}{2}i \end{bmatrix} $$

If we take the sum of squared differences of the components of these two vectors, we end up with

$$\left ( (\frac{1}{2} + \frac{1}{2}i) - (-\frac{1}{2} + \frac{1}{2}i) \right )^2 = 1$$

and 

$$\left ( (\frac{1}{2} + \frac{1}{2}i) - (\frac{1}{2} - \frac{1}{2}i) \right )^2 = -1$$

So the loss at this point would be zero, even though $|U(\theta)\rangle$ and $|\psi\rangle$ are clearly very different. 

To account for cases like this, I decided to change the loss function to

$$\sum_i | |U(\theta)\rangle_i - |\psi\rangle_i |^{2}$$

with $|a + bi|$ representing $a^2 + b^2$. This function avoids cases like the one above by assigning a positive value to every  difference $|U(\theta)\rangle_i - |\psi\rangle_i$.

### Solution

The solution can be executed by running all the code cells below, starting with the imports required:

```python
from qiskit import QuantumRegister
from qiskit import Aer, execute, QuantumCircuit
from qiskit.quantum_info import random_statevector
from qiskit.extensions import UnitaryGate
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

Next we define all the functions we need to construct the variational circuit - the noise parameters aren't really important now, but we'll come back to them later:

```python
def getRZGate(angle):
    """
    Returns a UnitaryGate that is equivalent to the
    RZ(angle) gate.
    """
    cos = np.cos(angle / 2)
    sin = np.sin(angle / 2)
    uGate = np.array(
        [[complex(cos, -sin), 0],
            [0, complex(cos, sin)]]
    )
    return UnitaryGate(uGate)


def oddBlock(qc, qreg, params, numQubits, noise):
    """
    Appends a layer of RX gates applied onto
    the qubits of qreg in the circuit qc.

    Normally distributed noise noise is added to the
    parameters used with the RX gates to simulate
    noisy hardware.
    """
    samples = np.random.normal(0, noise, numQubits)
    for i in range(numQubits):
        qc.rx(params[i] + samples[i], qreg[i])


def evenBlock(qc, qreg, params, numQubits, noise):
    """
    Appends a layer of RZ gates followed by an
    entanglement layer made up of CZ gates applied onto
    the qubits of qreg in the circuit qc.

    Normally distributed noise noise is added to the
    parameters used with the RZ gates to simulate
    noisy hardware.
    """
    samples = np.random.normal(0, noise, numQubits)
    for i in range(numQubits):
        qc.append(getRZGate(params[i] + samples[i]), [qreg[i]])
    for i in range(numQubits - 1):
        for j in range(i + 1, numQubits):
            qc.cz(qreg[i], qreg[j])


def circuit(qc, qreg, params, noiseVals):
    """
    Applies n alternating layers of RX and RZ + CZ
    gates onto the register qreg in the circuit qc,
    where n is the length of the second dimension of
    the array params.

    The values in noiseVal are used to simulate
    noise that increases with each layer in the circuit.
    """
    numQubits = len(params[0])
    for i in range(len(params) // 2):
        j = 2 * i
        noise = noiseVals[0] * (noiseVals[1] ** i)
        oddBlock(qc, qreg, params[j], numQubits, noise)
        evenBlock(qc, qreg, params[j + 1], numQubits, noise)
```

Now everything we need to compute the loss:

```python
def getStatevector(qc, backend):
    """
    Returns a statevector given a QuantumCircuit
    and the statevector_simulator backend.
    """
    job = execute(qc, backend=backend, shots=1)
    results = job.result().get_statevector()
    return results


def computeSquaredDistance(paramsPred, vector, backend, noiseVals):
    """
    Returns the value ∑ |fᵢ - vᵢ| ** 2, where fᵢ is
    the ith element of the output of the variational
    circuit, and vᵢ the ith element of the vector passed
    into the parameter vector.
    """
    qreg = QuantumRegister(len(paramsPred[0]))
    qc = QuantumCircuit(qreg)
    circuit(qc, qreg, paramsPred, noiseVals)
    pred = getStatevector(qc, backend)
    diff = pred - vector
    loss = 0
    for i in range(len(diff)):
        loss += diff[i].real ** 2 + diff[i].imag ** 2
    return loss, diff
```

These functions compute the gradients and update the parameter set for us:

```python
def computeGradients(paramsPred, vector, lossVec, backend, noiseVals):
    """
    Returns the partial derivatives of the loss function
    defined in computeSquaredDistance with respect to
    the parameter set paramsPred.
    """
    newParams = np.copy(paramsPred)
    gradients = np.zeros_like(paramsPred)
    for i in range(len(paramsPred)):
        for j in range(len(paramsPred[0])):
            newParams[i, j] += np.pi
            qreg = QuantumRegister(len(paramsPred[0]))
            qc = QuantumCircuit(qreg)
            circuit(qc, qreg, newParams, noiseVals)
            gradVec = getStatevector(qc, backend)
            grad = 0
            for k in range(len(gradVec)):
                grad += lossVec[k].imag * gradVec[k].imag
                grad += lossVec[k].real * gradVec[k].real
            gradients[i, j] = grad
            newParams[i, j] -= np.pi
    return gradients / 2


def updateParams(prevParams, params, grads, lr, momentum):
    """
    Updates params using the gradients passed
    into the grads param, the previous parameter values
    prevParams, and learning rate lr and
    gamma value momentum.
    """
    diff = params - prevParams
    newParams = params - grads * lr + diff * momentum
    return newParams, params
```

Now we can optimize the circuit parameters given a vector and a circuit structure:

```python
def trainCircuit(vector, numLayers, numQubits, backend, iters, noiseVals):
    """
    Optimizes the variation circuit to minimize the
    loss function computeSquaredDistance given a
    statevector vector. The structure of the variational
    circuit is determined by parameters numLayers and
    numQubits, and the number of optimization steps
    is given by iters. Learning is performed using SGD
    + Nesterov momentum.
    """
    params = np.random.sample((2 * numLayers, numQubits))
    prevParams = np.copy(params)
    lr = 0.02
    momentum = 0.9
    losses = []
    for i in range(iters):
        loss, lossVec = computeSquaredDistance(
            params, vector, backend, noiseVals
        )
        losses.append(loss)
        print('Iter {} Loss: {}'.format(i + 1, loss))
        diff = params - prevParams
        accParams = params + diff * momentum
        gradients = computeGradients(
            accParams, vector, lossVec, backend, noiseVals
        )
        params, prevParams = updateParams(
            prevParams, params, gradients, lr, momentum
        )
    return losses
```

Finally, we create a function that repeatedly calls *trainCircuit* while changing the number of layers used in the variational circuit, and execute it. We'll start off with a circuit that only has 1 layer, and go on adding layers until we have a circuit with 10. Each time, we'll perform 500 parameter updates:

```python
def getMinimumDistances(vector, numQubits, backend, layerList, noiseVals):
    """
    Returns a list of losses for circuits with layer
    depths given in layerList.

    The list noiseVals should contain the probabilities
    pPhase and pAmp to be used to simulate phase
    and amplitude damping, respectively.
    """
    lossesList = [trainCircuit(
        vector, i + 1, numQubits, backend, 500, noiseVals) for i in layerList]
    return lossesList
```

First, though, let's generate a random vector that we can use throughout the notebook:

```python
vector = random_statevector(16).data
vector
```

Great - now let's call *getMinimumDistances*:

```python
np.random.seed(0)
backend = Aer.get_backend('statevector_simulator')
noiseVals = [[-1, -1] for i in range(4)]
lossesList = getMinimumDistances(vector, 4, backend, range(10), noiseVals)
```

Let's look at the minimum loss for each circuit:

```python
minLosses = [i[-1] for i in lossesList]
for i in range(10):
    print('Layers : {} | Loss : {}'.format(i + 1, minLosses[i]))
```

### Result visualization

Now we can use the loss information to plot some graphs - here's one that shows how the minimum loss changes with the number of layers used in the variational circuit:

```python
x1 = [i + 1 for i in range(10)]
plt.figure(figsize=[16, 9])
plt.yscale('log')
plt.plot(x1, minLosses)
plt.plot(x1, minLosses, 'ro')
plt.xlabel('Layers')
plt.ylabel('Minimum loss')
plt.legend(['Minimum Loss'])
plt.title('Minimum loss v/s circuit layer depth')
plt.show()
```

Here's another that shows how the loss changes with the number of optimization iterations:

```python
plt.figure(figsize=[16, 9])
plt.yscale('log')
x2 = [i + 1 for i in range(500)]
for i in lossesList:
    plt.plot(x2, i)
legendList = ['{} Layers'.format(i + 1) for i in range(10)]
plt.xlabel('Optimization steps')
plt.ylabel('Loss')
plt.legend(legendList)
plt.title('Loss v/s optimization iterations')
plt.show()
```

We can see that the circuit performs relatively poorly when it is composed of only a few layers; this is to be expected, as the portion of the Hilbert space it can access is small. Increasing the number of layers in the variational circuit yields a marked improvement in performance - but we also see that the circuit with 7 layers performed better than the one with 8. 

This could be due to the fact that once the circuit is expressively rich enough and losses due to the circuit being unable to reach certain areas of the Hilbert space are reduced, the effects of other factors (like the initial parameter values) become more apparent.

Interestingly, we can tell from the second graph that the gradients for the circuits with fewer layers fell to zero very quickly - the circuits with 1, 2 and 3 layers experience very small reductions in loss after about 150 iterations. The gradients of the circuits with more than 4 layers all seem to be significant, hinting at further reductions in loss should the optimization procedure be performed for more than 500 iterations.

### Circuit performance with simulated noise

Finally, all the results so far have been obtained using an ideal classical simulator - what happens if we introduce some noise? At what level of noise do our results start to get affected? Do the results for circuits with different layer depths get affected differently?

The statevector simulator backend does not offer support for noise models, but we can still simulate the effects of quantum noise with a super simple noise model that adds $x \sim N(0, \sigma^2)$ to every parameter during circuit execution, with $\sigma$ being a function of the layer number currently being executed - this way we can account for the relaxation and dephasing times of the qubits we use, which result in errors becoming worse and more frequent as the circuit depth increases. Specifically, we use this formula to draw random values:

$$x \sim N(0, \sigma^2), \: \sigma = a * b^l$$

Here $a$ is a small constant (we'll use a default value of 0.01), and $b$ is a number greater than 1 (we use values 1.05, 1.1, and 1.15). $l$ represents the layer number currently being applied. Let's try this out with a couple different noise levels and a few different circuit layer depths - we'll use circuits with 5, 7, and 10 layers:

```python
plt.figure(figsize=[16, 9])
plt.yscale('log')
colorsList = ['b', 'r', 'g']
stylesList = ['-', '--', ':']
for i in range(3):
    for j in range(3):
        plt.plot(x2, noisyLossesList[i][j], colorsList[j] + stylesList[i])
legendList = ['{} Layers: Noise settings {}'.format(i + 1, noiseVals[j]) for j in range(3) for i in layers]
plt.legend(legendList)
plt.xlabel('Optimization steps')
plt.ylabel('Loss')
plt.title(
    'Loss v/s optimization iterations with simulated noise'
)
plt.show()
```

The first things we notice is that performance with noise is significantly worse than it is without it - another is that the level of noise introduced has definite effect on loss values. Circuit depth does not seem to have as great an effect as it did when we ran ideal simulations. 

When the noise levels were low, the circuit with 10 layers did best, but once noise levels were increased, the circuit with 7 layers performed better - this could be due to the difference in the standard deviation used for the distribution we draw the noise samples from: layers 8, 9, and 10 used normal distributions with standard deviations 0.2144 and 0.0306, 0.0236 and 0.0352, and 0.0259 and 0.0405 for noise settings \[0.01, 1.1\] and \[0.01, 1.15\], respectively, while layer 7 used standard deviations 0.01949 and 0.02660 for the same.

Another insight we can draw from the graphs are that this time the circuit with 10 layers converges to its final value very quickly (for example, around iteration 200 in the graph with noise settings \[0.01, 1.1\], and iteration 250 in the next) and remains largely stagnant after that, apart from fluctuations due to noise. On the other hand, the same graphs show that the loss for circuits with 5 and 7 layers remain sloping downwards at 500 iterations, hinting at possible further reductions in loss.

This could be due to the higher noise levels experienced by the circuit with 10 layers, which overpower the true gradient signal, while the noise in the 7 and 5 layer cases remain low enough to allow the optimizer to extract useful gradient information.

Simply put, the sweet spot seems to be around 7 layers - the circuit is deep enough to capture complex relationships between qubits, and can access a large portion of the Hilbert space, while still being small enough to avoid reductions in performance as great as circuits with 10 layers do.
    
### Next steps

Awesome!

The variational circuit solves task 1 pretty well - with a 10 layer circuit running on an ideal simulator we acheived a loss of order 1e-7 after optimizing for only 500 iterations. There are still a few things we could do to boost circuit performance and draw more accurate conclusions:

* Use Adam instead of SGD + Nesterov momentum for faster convergence.
* Run more circuits with different levels of noise for a better understanding of how noise tolerant the circuit is, and how its tolerance depends on the number of layers used.
* Simulate thermal relaxation using different $T_1$ and $T_2$ times for each qubit for more accurate noise modelling.
* Switch up the gates used in the circuit and compare results - I ran the circuit with $R_Y$ and $CX$ gates instead of $R_Z$ and $CZ$ ones, but the results are not reported here.
* Experiment with circuit structure - instead of two layers of single qubit gates and then one made up two qubit gates that serves as an entanglement layer, we could try using multiple layers of two qubit gates like the $XX$ and $ZX$ gates, as well as  their rotational variants.
